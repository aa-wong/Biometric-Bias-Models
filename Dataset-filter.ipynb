```python
import pandas as pd
```

```python
normalized_data = './data/norming_data.csv'
```

```python
df = pd.read_csv(normalized_data)
df.head()
```

```python
import os
from shutil import copy
```

```python
# target: Array of target directory names
# source_dir: directory where all target folders are located
# dest_dir: directory to place all copied files
# file_type: file type to copy

def copy_and_place_files_in_directory(targets, source_dir, dest_dir, file_type):
    directories = dest_dir.split('/')
    
    dest_path = '.'
    for level in directories:
        if level != '.':            
            dest_path += '/' + level
            if not os.path.exists(dest_path):
                os.makedirs(dest_path)

    for target in targets:
        target_dir = source_dir + '/' + target
        for file_name in os.listdir(target_dir):
            dest_file_path = dest_path + '/' + file_name
            if file_name.endswith(file_type) and not os.path.isfile(dest_file_path):
                source_file = target_dir + '/' + file_name
                copy(source_file, dest_path)
```

```python
field = 'Trustworthy'
```

```python
dataset = df[field]
low = dataset.quantile(0.25)
med = dataset.quantile(0.5)
high = dataset.quantile(0.75)
```

```python
non_data = df.loc[dataset < low]
non_data.head()
low_data = df.loc[(dataset > low) & (dataset < med)]
low_data.head()
med_data = df.loc[(dataset > med) & (dataset < high)]
med_data.head()
high_data = df.loc[dataset > high]
high_data.head()
```

```python
data_sets = [
    {
        "data": non_data.loc[non_data['Gender'] == 'F'],
        "name": "non_female"
    },
    {
        "data": non_data.loc[non_data['Gender'] == 'M'],
        "name": "non_male"
    },
    {
        "data": low_data.loc[low_data['Gender'] == 'F'],
        "name": "low_female"
    },
    {
        "data": low_data.loc[low_data['Gender'] == 'M'],
        "name": "low_male"
    },
    {
        "data": med_data.loc[med_data['Gender'] == 'F'],
        "name": "med_female"
    },
    {
        "data": med_data.loc[med_data['Gender'] == 'M'],
        "name": "med_male"
    },
    {
        "data": high_data.loc[high_data['Gender'] == 'F'],
        "name": "high_female"
    },
    {
        "data": high_data.loc[high_data['Gender'] == 'M'],
        "name": "high_male"
    },
]
```

```python
for data in data_sets:
    targets = data["data"]
    destination_dir = './data/dataset/' + field.lower() + '/' + data["name"]
    copy_and_place_files_in_directory(targets["Target"], 
                                      './data/images', 
                                      destination_dir, 
                                      '.jpg')
```
